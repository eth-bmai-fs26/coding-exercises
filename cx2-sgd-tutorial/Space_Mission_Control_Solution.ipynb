## 📡 Exercise 1: The Robbins-Monro Update

In Mission 1, you found the minimum of the terrain manually.
But we cannot pilot every landing by hand.
We need an automated system.

We will use an iterative update rule called the **Robbins-Monro algorithm**.
It works by taking small steps toward the target.
If we are far away, we take bigger steps.
As we get closer, we slow down.

The update rule is:

$$w_t \leftarrow w_{t-1} + \alpha_t \cdot (w^* - w_{t-1})$$

Where:
* $w_{t-1}$ is our current position.
* $w^*$ is the target (the minimum you found in the game).
* $\alpha_t$ is our "step size" (how aggressive we are).

**Your Task:** Complete the code below to implement this update loop.

```python
# --- Configuration ---
w_star = -3.0      # The target minimum (matches the game function w^2 + 6w + 10)
w = 0.0            # Starting position (initial guess)
alpha = 0.1        # Step size (learning rate)
iterations = 20    # Number of updates

print(f"Starting descent from w = {w}...")

# --- The Update Loop ---
for t in range(iterations):
    # Update 'w' using the formula: w_new = w_old + alpha * (target - w_old)
    w = w + alpha * (w_star - w)

    # Print the current position
    print(f"Time {t+1}: Current w = {w:.4f}")

print(f"\nFinal Position: {w:.4f} (Target: {w_star})")
```

## 📡 Exercise 2: Coding the Slope Sensor

In Mission 3, you learned that the slope can be estimated by taking two sensor readings and finding the difference.
This is the **Finite Difference Method**.

We define a small window of size $\varepsilon$:
* **Left point ($l$):** $w - \varepsilon/2$
* **Right point ($r$):** $w + \varepsilon/2$

The slope is:

$$\nabla_\varepsilon L(w) = \frac{L(r) - L(l)}{\varepsilon}$$

You are given a "black box" function $L(w)$ that returns the loss for any position $w$.
You don't know the formula inside, but you can query it.

**Your Task:** Implement the function `calculate_finite_difference` to return the slope.

```python
def L(w):
    """The mysterious terrain function (Oracle)."""
    return w**2 - 4*w + 7

def calculate_finite_difference(w, epsilon=0.1):
    """
    Calculates the slope of L at point w using a window of size epsilon.
    """
    # Calculate the left point 'l' and right point 'r'
    l = w - (epsilon / 2)
    r = w + (epsilon / 2)

    # Get the height at these points using the function L()
    height_l = L(l)
    height_r = L(r)

    # Calculate the slope (rise over run)
    slope = (height_r - height_l) / epsilon

    return slope

# --- Test your sensor ---
current_w = 0.0
slope = calculate_finite_difference(current_w)
print(f"At w={current_w}, the estimated slope is: {slope:.4f}")
print("If the slope is negative, we should move Right. If positive, move Left.")
```

## 📡 Exercise 3: Building the 1D Autopilot

In Mission 5, you saw Gradient Descent in action.
The algorithm calculates the gradient at our current position and moves in the opposite direction.

$$w_{new} = w_{old} - \eta \cdot \nabla L(w_{old})$$

* If the slope is positive (uphill to the right) $\to$ we subtract (move left).
* If the slope is negative (downhill to the right) $\to$ we add (move right).

**Your Task:** Complete the gradient descent loop below.

```python
def grad_L(w):
    """The Gradient function (Derivative).
    For L(w) = w^2 - 4w + 7, the derivative is 2w - 4.
    """
    return 2*w - 4

# --- Configuration ---
w = 0.0          # Initial position
eta = 0.1        # Learning rate (step size)
iterations = 15  # Total steps

print(f"Starting Autopilot from w = {w}...")

for t in range(iterations):
    # 1. Calculate the gradient at the current position
    gradient = grad_L(w)

    # 2. Update 'w' using the Gradient Descent rule
    # Formula: w_new = w_old - eta * gradient
    w = w - eta * gradient

    print(f"Step {t+1}: Gradient = {gradient:.2f}, New w = {w:.4f}")

print("Touchdown complete.")
```
