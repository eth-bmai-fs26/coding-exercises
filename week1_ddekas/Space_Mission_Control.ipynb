* **Left point ($l$):** $w - \varepsilon/2$
* **Right point ($r$):** $w + \varepsilon/2$

The slope ($\nabla_\varepsilon L$) is simply the "rise over run":

$$\nabla_\varepsilon L(w) = \frac{L(r) - L(l)}{\varepsilon}$$

## **🔬 Game 4: The Precision Limit**

**The Situation:** In Game 3, we used a window of size $\varepsilon$ to estimate the slope. But is that estimate perfect?

If our window is huge (like looking at a mountain from space), we miss the details and our slope calculation is wrong. If we shrink the window, our estimate becomes more accurate.

**The Concept:** As we shrink the window size down to zero ($\varepsilon \to 0$), our "estimated slope" stops changing and converges to a single, precise number.

This magical number is the **Gradient** ($\nabla L(w)$).

$$\nabla L(w) = \lim_{\varepsilon \to 0} \frac{L(w + \varepsilon/2) - L(w - \varepsilon/2)}{\varepsilon}$$

## **📉 Game 5: Engaging Autopilot (Gradient Descent)**

2. **The Navigation:** We know how to calculate the exact direction downhill (The Gradient from Game 4).

It is time to combine them into the ultimate algorithm: **Gradient Descent**.

**The Algorithm:** Instead of guessing the direction, we calculate the gradient $\nabla L(w)$ at our current position. Since the gradient points "uphill," we simply move in the opposite direction to go "downhill."

$$w_{new} = w_{old} - \alpha \cdot \nabla L(w_{old})$$

* If the slope is positive (uphill to the right) $\to$ we subtract (move left).
* If the slope is negative (downhill to the right) $\to$ we add (move right).
